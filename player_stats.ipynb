```python
#import dependancies
import pandas as pd
from sqlalchemy import create_engine
```

```python
#connect to database
rds_connection_string = "postgres:root123@localhost:5432/NBA_DB"
engine = create_engine(f'postgresql://{rds_connection_string}')
engine.table_names()
```

```python
#loading file
csv = "Resources/playerstats2018-2019.csv"

player_stats = pd.read_csv(csv).dropna()
player_stats.head()
```

```python
# rename the columns based on other tables
player_stats = player_stats.rename(columns={"Name":"full_name","Team": "team_id","Points": "PTS","Blocks": "BLK","Steals": "STL","Assists": "AST",
                                            "Rebounds": "REB","FT%": "FT_PCT","FTA": "FTA","FG3%": "FG3_PCT","FG3A": "FG3A",
                                            "FG%": "FG_PCT","FGA": "FGA","MP": "MIN","G": "GP"})
player_stats.head(10)
```

```python
#drop the ones that are not need
player_stats = player_stats.reset_index().drop(columns=["index","Height","Weight","Age","Salary","PER","OWS","WS","WS48","DWS"], axis=1)
player_stats.head(10)
```

```python
# insert the season_id
player_stats.insert(1, "season_id", "2018-19")
player_stats.head()
```

```python
# #loop for all the teams
team_list = player_stats['team_id'].values.tolist()
team_id_list = []

for x in range(len(team_list)):
    try:
        id_df = pd.read_sql_query("SELECT team_id FROM team WHERE full_name = '" + player_stats["team_id"][x] + "'", con=engine)
        id_list = id_df.values.tolist()
        #print(id_list[0][0])
        team_id_list.append(id_list[0][0])
    except:
        print("NAN")
```

```python
len(team_id_list)
```

```python
# dropping the team
player_stats = player_stats.drop(columns=['team_id'])
player_stats.head(10)
```

```python
# putting the 3 letter team id in
player_stats.insert(1, "team_id", team_id_list)
player_stats
```

```python
# selecting the player table from SQL
player_table = pd.read_sql_query("SELECT player_id, full_name FROM player", con=engine)
player_table.head()
```

```python
#merging the 2 tables together
result = pd.merge(player_table, player_stats, how='inner', on=["full_name"])
result
```

```python
#final dataframe
final_player = result.drop(columns=["full_name"])
final_player
```

```python
#to SQL
final_player.to_sql(name='player_stats', con=engine, if_exists='append', index=False)
```
