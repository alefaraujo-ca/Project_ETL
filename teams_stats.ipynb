```python
from bs4 import BeautifulSoup as bs
from splinter import Browser
import time
import pandas as pd
```

# Scrape teams stats from NBA homepage

```python
executable_path={'executable_path':'chromedriver.exe'}

browser = Browser('chrome', **executable_path, headless=False)

url = 'https://ca.global.nba.com/statistics/teamstats/'
browser.visit(url)
```

```python
# Using Inspect, we found the region of the HTML page that contain the table with the stats information
```

```python
html = browser.html
soup = bs(html, 'lxml')
table=soup.find('div', class_='nba-stat-table__overflow')


rows=table.find_all('tr')
pre_df=[]
columns=['rank', 'team', 'G', 'FG_PCT','3P_PCT','FT_PCT','OREB','DREB','PPG','RPG','APG','TO','SPG','BPG','PF']
```

```python
# Loop through the tags and store the data stats into a dataframe
```

```python
pre_df=[]
for row in rows[1:]:
    team_dict={}
    cells=row.find_all('td')

    for i, cell in enumerate(cells[0:15]):
        team_dict[columns[i]]=cell.text.strip()
        print(i, cell.text.strip())
    
    pre_df.append(team_dict)
    

df_team_stats=pd.DataFrame(pre_df)
```

```python
# Rename the column team to full_name, as it is the field created at our database
df_team_stats = df_team_stats.rename(columns={'team':'full_name'})
df_team_stats.to_csv('./Resources/team_stats.csv')
```

```python
from sqlalchemy import create_engine
from datetime import datetime
```

```python
rds_connection_string = "postgres:root123@localhost:5433/NBA_DB"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
# Retrieve the teams we have stored at database, so we can compare with the results scraped
df_teams = pd.read_sql_query("select team_id, full_name from team", con=engine)
```

```python
# Merge the stats df with the teams df to find the id for each team
```

```python
merge_result = pd.merge(df_teams, df_team_stats, how='inner', on=['full_name'])
```

```python
# Drop the columns we do not need
df_teams_stats_db = merge_result.drop(columns=['rank', 'full_name'])
```
